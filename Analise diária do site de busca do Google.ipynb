```python
!pip install pytrends
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

# Criando um DataFrame dos 10 paises com a palavra Machine Learning no site de busca do Google

```python
trends.build_payload(kw_list = ['Machine Learning'])
dataset = trends.interest_by_region()
dataset = dataset.sort_values(by="Machine Learning", ascending=False)
dataset = dataset.head(20)
print(dataset)
```

```python
dataset.reset_index().plot(x="geoName", 
                        y="Machine Learning", 
                        figsize=(15,12), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
dataset = TrendReq(hl='en-US', tz=360)
dataset.build_payload(kw_list=['Machine Learning'])
dataset = dataset.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 12))
dataset['Machine Learning'].plot()
plt.style.use('fivethirtyeight')
plt.title('O numero total de buscas no Google sobre Machine Learning', 
          fontweight='bold')
plt.xlabel('Ano')
plt.ylabel('Contagem')
plt.show()
```
